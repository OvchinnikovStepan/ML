Начнём работать со значениями

Переведём IMDB Votes в численный формат

```python
data['IMDB Votes']=data['IMDB Votes'].str.replace(',','').astype(float)
data['IMDB Votes'].unique()
```

Проверили

Посмотрим какие значения у нас пропущены

```python
(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)
```

Пропуски в дискретных данных заменим на "Unknown

Пропуски в значениях рейтинга на Tomatometer заполним округлёнными значениями с IMDB, на основе этого же рейтинга определим статус фильма "Fresh"
или "Rotten", заодно убрав дубликаты, возбмём число критиков как 60 (ближайшее число кратное 10 отступая от медианы к среднему) и распределим их
голоса на основе рейтинга. "Audience (Count,Rating,Status)" так же заполним на осноме IMDB.

В пропуски даты релиза положим 1 июля года выпуска

```python
columns_to_replace =['Critic Consensus', 'Audience Status', 'Authors','Directors','Movie Info','Actors','Production Company','Content Rating','Genres']

for row in columns_to_replace:
    data[row] = data[row].fillna('Unknown')

data['Tomatometer Rating']=data['Tomatometer Rating'].fillna(round(data['IMDB Rating'])*10)
data['Audience Rating']=data['Audience Rating'].fillna(round(data['IMDB Rating'])*10)

data.loc[data['Tomatometer Rating']>=60,'Tomatometer Status']="Fresh"
data.loc[data['Tomatometer Rating']<60,'Tomatometer Status']="Rotten"

data.loc[data['Audience Rating']>=60,'Audience Status']="Upright"
data.loc[data['Audience Rating']<60,'Audience Status']="Spilled"

data['Tomatometer Count']=data['Tomatometer Count'].fillna(60)
data['Tomatometer Fresh Critics Count']=data['Tomatometer Fresh Critics Count'].fillna(data['Tomatometer Rating']*0.6)
data['Tomatometer Top Critics Count']=data['Tomatometer Top Critics Count'].fillna(round(data['Tomatometer Fresh Critics Count']*0.2))
data['Tomatometer Rotten Critics Count']=data['Tomatometer Rotten Critics Count'].fillna(60-data['Tomatometer Fresh Critics Count'])

data['Audience Count']=data['Audience Count'].fillna(data['IMDB Votes'])

data['Original Release Date']=data['Original Release Date'].fillna(data['Year of Release'].astype(str)+"-07-01")
```

```python
data.describe()
```

В пропуски в выпусках на стриминговые сервисы положим 1 декабря 2016 (Красивая дата близкая к медиане)

```python
data['Streaming Release Date']=data['Streaming Release Date'].fillna('2016-01-01')
```

Преобразуем в формат даты соответсвующие поля

```python
data ['Streaming Release Date'] = pd.to_datetime(data['Streaming Release Date'],format = '%Y-%m-%d')
data ['Original Release Date'] = pd.to_datetime(data['Original Release Date'],format = '%Y-%m-%d')
```

В тех случаях когда дата выпуска на стриминговые сервисы раньше чем просто релиза заменим дату выпуска н дату релиза

```python
data.loc[data['Streaming Release Date']<data['Original Release Date'],'Streaming Release Date']= \
data.loc[data['Streaming Release Date']<data['Original Release Date']]['Original Release Date']
```

НАрисуем гистограммы

```python
columns = ['IMDB Rating','IMDB Votes','Original Release Date','Movie Time']

for column in columns:
    data.hist(column)
```

Заметим что 'IMDB Votes' имеет значительные выбросы и 'Movie Time' незначительные. Офильтруем данные

```python
outlier = data[['IMDB Votes','Movie Time']]
Q1 = outlier.quantile(0.25)
Q3 = outlier.quantile(0.75)
IQR = Q3-Q1
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
index_list = list(data_filtered.index.values)
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

Мы отфильтровали данные и сохранили их в отдельной переменной

```python
filtered_outlier = data_filtered[['IMDB Votes','Movie Time']]

for col in filtered_outlier:
    data_filtered.hist(col, bins=100)
```

Несмотря на удаление выбрасов IMDB Votes гистограмма всё равно сильно смещенна влево, что сведетельствует о большом числе фильмов с малым числом оценок. Гистограмма длительности фильма же стала более симметричной и из рассмотрения были удалены фильмы длительностью более 3 часа.

Найдём средний рейтинг IMDB для фильмов каждого года

```python
avg_rating = data_filtered.pivot_table(index='Year of Release', values='IMDB Rating', aggfunc='mean')
avg_rating.sort_values(ascending=False ,by='Year of Release')
```

Теперь мы имеем сводк по годам, представим на графе

```python
avg_rating=avg_rating.reset_index()
avg_rating.plot(grid=True, figsize=(30, 3), kind = 'bar', x='Year of Release',y='IMDB Rating')
```

Давайте посмотрим как на оценку фильма влияет его длительность и количество оценок

```python
sns.lmplot(x="IMDB Votes", y="IMDB Rating", data=data_filtered)
plt.title('Votes-Rating')
sns.lmplot(x="Movie Time", y="IMDB Rating", data=data_filtered)
plt.title('Duration-Rating')
```

МЫ можем наблюдать что большое количество оценок имеют прежде всего популярные и высокооценённые фильмы, данные коррелируют. Так же можно заметить, что более продолжительные фильмы имеют немного больший шанс понравится зрителям. Вычислим корреляцию

```python
votes_rating = data_filtered[["IMDB Votes","IMDB Rating"]]
corr_votes = votes_rating.corr()
dur_rating = data_filtered[['Movie Time',"IMDB Rating"]]
corr_dur = dur_rating.corr()
```

```python
round(corr_votes,3)
```

```python
round(corr_dur,3)
```

Насколько показатели зависимы можно видеть на коэфиценте корреляции
