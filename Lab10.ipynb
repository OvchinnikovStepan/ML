```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
data
```

```python
data.rename(columns = {'v1': 'status', 'v2': 'incoming_text'}, inplace = True)
data['status'] = data['status'].map({'ham': 1, 'spam': 0})
```

```python
import re
```

```python
data['processed_text'] = data['incoming_text'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x).lower())
```

```python
data
```

Токенизация

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
data['tockens'] = data['processed_text'].apply(lambda x: [token for token in nlp(x)])
```

```python
data
```

Исключение стоп-слов

```python
stopwords = nlp.Defaults.stop_words
```

```python
data['tockenlist'] = data['tockens'].apply(lambda x: [token for token in x if token.lower_  not in stopwords])
```

```python
data
```

Лемматизация

```python
data['lem_tockenlist'] = data['tockenlist'].apply(lambda x: [token.lemma_ for token in x])
```

```python
data
```

Создадим список строк

```python
strings = []
for i in data['lem_tockenlist']:
    text =' '.join([str(row) for row in i])
    strings.append(text)
```

```python
strings
```

Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(ngram_range=(1,4),min_df=3)
```

```python
bow = count_vectorizer.fit_transform(strings)
bow.shape
```

```python
bow.todense()
```

TD-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,4),min_df=3)
tfidf = tfidf_vectorizer.fit_transform(strings)
tfidf_vectorizer.vocabulary_
```

```python
tfidf.shape
```

Понижение размерности

```python
from sklearn.feature_selection import SelectKBest
```

```python
Y=data['status']
```

```python
skb = SelectKBest(k=80)
X_td = skb.fit_transform(tfidf,Y)
X_td.shape
```

```python
skb = SelectKBest(k=80)
X_bag = skb.fit_transform(bow,Y)
X_bag.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
x_train_bow, x_test_bow, y_train_bow, y_test_bow = train_test_split(X_bag,Y, test_size=0.2)
x_train_td, x_test_td, y_train_td, y_test_td = train_test_split(X_td, Y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
x_train_bow.toarray()
```

```python
knn.fit(x_train_bow.toarray(),y_train_bow)
y_pred_bow = knn.predict(x_test_bow.toarray())
print(classification_report(y_test_bow, y_pred_bow))
```

```python
knn.fit(x_train_td.toarray(),y_train_td)
y_pred_td = knn.predict(x_test_td.toarray())
print(classification_report(y_test_td, y_pred_td))
```

## Тематическое моделирование

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
lda_bow = LatentDirichletAllocation(n_components=5, max_iter=20, learning_method='online', learning_offset=50)
```

```python
lda_bow_ = lda_bow.fit_transform(bow)
```

```python
for i, comp in enumerate(lda_bow.components_):
    vocab_comp = zip(count_vectorizer.get_feature_names_out(), comp)
    sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:10]
    print("Содержание темы "+str(i+1)+": ")
    for t in sorted_words:
        print(t[0],end=" ")
    print("\n")
```

```python
lda_tfidf = LatentDirichletAllocation(n_components=5, max_iter=20, learning_method='online', learning_offset=50)
```

```python
lda_tfidf_ = lda_tfidf.fit_transform(tfidf)
```

```python
for i, comp in enumerate(lda_tfidf.components_):
    vocab_comp = zip(count_vectorizer.get_feature_names_out(), comp)
    sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:10]
    print("Содержание темы "+str(i+1)+": ")
    for t in sorted_words:
        print(t[0],end=" ")
    print("\n")
```
